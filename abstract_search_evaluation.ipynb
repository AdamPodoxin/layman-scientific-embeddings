```python
from pathlib import Path
from abstract_search_evaluation import get_scores
```

```python
MODELS_PATH = Path("models")
```

```python
def print_model_scores(model_path: str | Path):
    scores = get_scores(model_path)
    print(model_path)
    print(f"Full match score: {scores["perfect_match_score"] * 100:.2f}%")
    print(f"In top results score: {scores["in_top_results_score"] * 100:.2f}%")
```

```python
for model_path in MODELS_PATH.iterdir():
    print_model_scores(model_path)
    print("")
```

```python
print_model_scores("allenai/scibert_scivocab_uncased")
```
