```python
from pathlib import Path
from keyword_search_evaluation import get_scores
```

```python
MODELS_PATH = Path("models")
```

```python
def print_model_scores(model_path: str | Path):
    scores = get_scores(model_path)
    print(model_path)
    print(f"Full match score: {scores["perfect_match_score"] * 100:.2f}%")
    print(f"Related keywords score: {scores["related_keyword_score"] * 100:.2f}%")
```

```python
for model_path in MODELS_PATH.iterdir():
    print_model_scores(model_path)
    print("")
```

```python
print_model_scores("allenai/scibert_scivocab_uncased")
```

```python
import pandas as pd
```

```python
scores_for_select_model = get_scores("models/jargon-layman-finetuned-1.00")
search_results: list[list[dict[str, float]]] = scores_for_select_model["search_results"]
search_df: pd.DataFrame = scores_for_select_model["search_df"]
```

```python
results_df = pd.DataFrame(data={
    "query_index": range(len(search_results)),
    "result_id": [
        [result["corpus_id"] for result in single_results]
        for single_results in search_results
    ]
}).explode("result_id")
results_df.head()
```

```python
results_df["layman"] = results_df["query_index"].apply(lambda i: search_df["layman"].iloc[i])
results_df["jargon"] = results_df["result_id"].apply(lambda i: search_df["jargon"].iloc[i])
results_df.head()
```
