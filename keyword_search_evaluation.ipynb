```python
from pathlib import Path
from keyword_search_evaluation import get_scores
```

```python
MODELS_PATH = Path("models")
```

```python
def print_model_scores(model_path: str | Path):
    scores = get_scores(model_path)
    print(model_path)
    print("Full match score:", scores["perfect_match_score"])
    print("Related keywords score:", scores["related_keyword_score"])
```

```python
for model_path in MODELS_PATH.iterdir():
    print_model_scores(model_path)
    print("")
```

```python
print_model_scores("allenai/scibert_scivocab_uncased")
```
